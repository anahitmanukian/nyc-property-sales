```python
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path().resolve().parent
CLEAN_PATH = PROJECT_ROOT / 'data' / 'clean' / 'clean.csv'

df = pd.read_csv(CLEAN_PATH)
```

```python
df.info()
```

```python
corr_cols =['PRICE_PER_SQFT',
            'SALE PRICE',
            'YEAR BUILT',
            'RESIDENTIAL UNITS',
            'COMMERCIAL UNITS','LAND SQUARE FEET','GROSS SQUARE FEET']

plt.figure(figsize=(8,6))
sns.heatmap(df[corr_cols].corr(),
            annot=True,
            cmap="coolwarm",   #  change this
            fmt=".2f")

plt.show()
```

```python
sns.boxplot(data=df, x='SALE YEAR', y='SALE PRICE')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.boxplot(data=df,
            x='SALE MONTH',
            y='SALE PRICE')

plt.xticks(rotation=45)
plt.show()
```

```python
monthly_avg = df.groupby('SALE MONTH')['SALE PRICE'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=monthly_avg,
            x='SALE MONTH',
            y='SALE PRICE')

plt.xticks(rotation=45)
plt.show()
```
